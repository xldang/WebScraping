```python
from bs4 import BeautifulSoup
import requests
import re
import time
```

### 深度优先搜索
**抓取所有内链**

```python
def DFS_scraping(url,match):
    start=time.time()
    webStack=[url]
    doneSet=set()
    while webStack!=[]:
        ##
        if time.time()-start>100:
            break
        ##
        u=webStack.pop()
        try:
            req=requests.get(u)
        except:
            doneSet.add(u)
            continue
        B_req=BeautifulSoup(req.content,'html.parser')
        #match=re.compile('fallincloud.com\/\?')
        print("Scraping... %s"%u)
        for link in B_req.find_all('a',href=match):
            if 'href' in link.attrs:
                if link['href'].startswith('http'):
                    if link['href'] not in doneSet:
                        print(link['href'])
                        doneSet.add(link['href'])
                        webStack.append(link['href'])
    end=time.time()
    
    return doneSet,end-start
```

```python
url='http://fallincloud.com/'
match=re.compile('fallincloud.com\/\?')
allUrl,cost=DFS_scraping(url,match)
```
