        </th>
        <th>4.99

        <th>Image.png
        </th>
    </tr>
</table>

**tr 是子标签（children）也是后代标签(descendant)，但th只是后代标签**

```python
url2='http://www.pythonscraping.com/pages/page3.html'
req=requests.get(url2)
B_req=BeautifulSoup(req.content,'html.parser')
```

```python
c=0
for child in B_req.find('table',{'id':'giftList'}).children:
    if c<3:
        print(child)
        c+=1
```

```python
c=0
for des in B_req.find('table',{'id':'giftList'}).descendants:
    if c<3:
        print(des)
        c+=1
```

### 2.2 兄弟标签

**next_siblings()** 获取一个标签的兄弟标签，
不会包含这个标签本身

```python
B_req.find('table',{'id':'giftList'}).tr
```

```python
for sib in B_req.find('table',{'id':'giftList'}).tr.next_siblings:
    print(sib)
```

### 2.3 正则表达式

只匹配 ../img/gifts/下的路径

```python
matcher=re.compile('\.\.\/img\/gifts\/img.*\.jpg')
for img in B_req.find_all('img',{'src':matcher}):
    print(img)
    print('—',img.attrs)
    print('——',img.attrs['src'])
```

## 3. 编写网络爬虫

**之所以称之为爬虫，就是因为它能顺着超链接递归爬取**

```python
url3='http://fallincloud.com/'
req=requests.get(url3)
B_req=BeautifulSoup(req.content,'html.parser')
```

```python
hrefs=set()
for link in B_req.find_all('a',href=re.compile('\/\?')):
    if 'href' in link.attrs:
        hrefs.add(link['href'])
hrefs
```

```python
s= 'http://fallincloud.com/?cat=55'
m=re.compile('fallincloud.com\/\?')
a=m.findall(s)
a
```

```python
def DFS_scraping(url):
    import 
    webStack=[url]
    doneSet=set()
    while webStack!=[]:
        u=webStack.pop()
        try:
            req=requests.get(u)
        except:
            doneSet.add(u)
            continue
        B_req=BeautifulSoup(req.content,'html.parser')
        match=re.compile('fallincloud.com\/\?')
        for link in B_req.find_all('a',href=match):
            if 'href' in link.attrs:
                if link['href'] not in doneSet:
                    print(link['href'])
                    doneSet.add(link['href'])
                    webStack.append(link['href'])
    return doneSet
```

```python
url3='http://fallincloud.com/'        
DFS_scraping(url3)
```
